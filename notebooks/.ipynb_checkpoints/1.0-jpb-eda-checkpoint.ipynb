```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

pd.pandas.set_option('display.max_columns', None)

import warnings
warnings.simplefilter(action='ignore')
```

```python
lending_data = pd.read_csv('lending_data.csv')

print(lending_data.shape)
lending_data.head()
```

## Clean Features

```python
# look for features that have % signs in their strings
percentage_feats = [feat for feat in lending_data.select_dtypes(include=['O'])
                    if lending_data[feat].str.contains('%').any()]

percentage_feats
```

```python
# remove percentage signs, convert to float dtypes
def remove_percentage_sign(df, feat):
    df = df.copy()
    df[feat] = df[feat].str.replace(r'%', '').astype('float')
    return df

for feat in ['int_rate', 'revol_util']:
    lending_data = remove_percentage_sign(lending_data, feat)
```

## Split Data

```python
lending_train, lending_val, rate_train, rate_val = train_test_split(
    lending_data, lending_data['int_rate'], test_size=0.1, random_state=12)
```

## Missing Values

```python
def feats_with_na_above_thresh(df, min_thresh, max_thresh=1.0):
    missing_feats = [feat for feat in df.columns if 
                     (df[feat].isnull().sum() / df.shape[0]) > min_thresh and
                     df[feat].isnull().sum() / df.shape[0] <= max_thresh]
    return df[missing_feats].isnull().mean().sort_values(ascending=False)
    
feats_with_na_above_thresh(lending_train, min_thresh=0.05)
```

When we preprocess, we will remove features that have greater than 90% missing; because we have plenty of data points, we will remove the instances that have less than 5% missing.

### Relationship between missing & interest rate

```python
def compare_na_value(df, feat):
    df = df.copy()
    df[feat] = np.where(df[feat].isnull(), 'missing', 'not missing')
    df.groupby(feat)['int_rate'].median().plot.bar()
    plt.ylabel('Interest Rate')
    plt.title(feat)
    plt.show()

# run function on features w/ < 90% missing & > 5% missing
na_feats_lt_90_and_grt_5 = feats_with_na_above_thresh(lending_train, 0.05, 0.9).index
for feat in na_feats_lt_90_and_grt_5:
    compare_na_value(lending_train, feat)
```

With these features, there does not exist much difference in interest rates between missing & not missing data points; so, this lack of difference suggests these predictors are not very important.

## Numerical Variables

```python
num_feats = [feat for feat in lending_train.columns if 
    lending_train[feat].dtypes != 'O']

print(f'Number of numerical variables: {len(num_feats)}')
lending_train[num_feats].head()
```

### Discrete Variables

```python
discrete_feats = [feat for feat in num_feats if len(
    lending_train[feat].unique()) < 20 and feat not in ['member_id', 'url', 'desc']]

print(f'Number of discrete variables: {len(discrete_feats)}')
lending_train[discrete_feats].head()
```

```python
def rate_to_discrete(df, feat):
    df = df.copy()
    df.groupby(feat)['int_rate'].median().plot.bar()
    plt.title(feat)
    plt.ylabel('Median Interest Rate')
    plt.show()
    
for feat in discrete_feats:
    if feat not in ['sec_app_inq_last_6mths', 
                    'sec_app_mort_acc', 
                    'sec_app_chargeoff_within_12_mths', 
                    'sec_app_collections_12_mths_ex_med',
                    'deferral_term',
                    'hardship_length']:
        rate_to_discrete(lending_train, feat)
```

There tends to be a relationship between the variables and interest rates, but this relationship is not always monotonic.

### Continuous Variables

```python
cont_feats = [
    feat for feat in num_feats if feat not in discrete_feats
        and feat not in ['member_id', 'url', 'desc']
]

print(f'Number of continuous features: {len(cont_feats)}')
```

### Categorical Variables

```python
cat_feats = [feat for feat in lending_train.columns if lending_train[feat].dtypes == 'O']

print(f'Number of categorical features: {len(cat_feats)}')
```

```python
# observe the cardinality
lending_train[cat_feats].nunique().sort_values(ascending=False)
```

Some of the high cardinality features will cause problems; *emp_title* will not provide any predictive power unless we can bin it somehow, *zip_code* and others can be problematic too. We may need to tackle some of these predictors during preprocessing.

**Rare Labels**

We will observe if any of the low cardinality features have labels that are present only in samll number of instances.

```python
def observe_rare_labels(df, feat, tgt, rare_thresh):
    df = df.copy()
    cat_perc = df.groupby(feat)[tgt].count() / df.shape[0]
    return cat_perc[cat_perc < rare_thresh]

# print categories of features that are present in less than 1% of instances
for feat in cat_feats:
    if feat not in [
        'id', 'emp_title', 'zip_code', 'earliest_cr_line', 'sec_app_earliest_cr_line'
    ]:
        print(observe_rare_labels(lending_train, feat, 'int_rate', 0.01))
        print()
```

There are quite a few features that show multiple labels that are present in less than 1% of the data points. We will deal with these rare labels in the preprocessing step.

We will explore relationship between the categories of different features and the interest rate.

```python
for feat in cat_feats:
    if feat not in [
        'id', 'emp_title', 'zip_code', 'earliest_cr_line', 'sec_app_earliest_cr_line'  
    ]:
        rate_to_discrete(lending_train, feat)
```

Some of these categories provide information on interest rates, but some of these predictors do not provide much signal, *emp_length*, *addr_state*, some of the date features, and *hardship_type*, which does not provide any variance.
